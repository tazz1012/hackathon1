```python
df = pd.read_csv(file_path)
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df, x='States/UTs', y='Attempt to commit Murder', color='blue')
plt.xticks(rotation = 90)
plt.title('Attempts to commit Murder by States/UTs')
plt.xlabel('States/UTs')
plt.ylabel('Attempts to commit Murder') 
plt.grid(True)
plt.show()
```

```python
data_grouped = df.groupby('States/UTs')['Attempt to commit Murder'].sum().reset_index()
plt.figure(figsize=(15,10))
sns.barplot(data=data_grouped, x='States/UTs', y='Attempt to commit Murder', color='blue')
plt.title('Attempts to Commit Murder Over Years')
plt.xlabel('States/UTs')
plt.ylabel('Attempts to Commit Murder')
plt.xticks(rotation=45)  
plt.grid(True)
plt.tight_layout()  
plt.show()
```

```python
numeric_columns = ['Robbery', 'Rape', 'Attempt to commit Rape'] 
df_numeric = df[numeric_columns]
sns.pairplot(df_numeric)
plt.suptitle('Pair Plot of Crime Data', y=1.02)  
plt.show()
```

```python
column_names = ['States/UTs', 'Robbery', 'Rape', 'Attempt to commit Rape']
fig, axs = plt.subplots(2, 2, figsize=(15, 12))

sns.boxplot(data=df, x='States/UTs', y='Robbery', ax=axs[0, 0])
axs[0, 0].set_title('Robbery by States/UTs')
axs[0, 0].tick_params(axis='x', rotation=90)

sns.boxplot(data=df, x='States/UTs', y='Rape', ax=axs[0, 1])
axs[0, 1].set_title('Rape by States/UTs')
axs[0, 1].tick_params(axis='x', rotation=90)

sns.boxplot(data=df, x='States/UTs', y='Attempt to commit Rape', ax=axs[1, 0])
axs[1, 0].set_title('Attempt to Commit Rape by States/UTs')
axs[1, 0].tick_params(axis='x', rotation=90)
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
column_names = ['Attempt to commit Murder', 'States/UTs', 'District']

fig, axs = plt.subplots(1, 1, figsize=(12, 8))

df[column_names[1]].value_counts().plot(kind='bar', ax=axs, color='skyblue')
axs.set_title(f'{column_names[1]} Distribution')
axs.set_xlabel(column_names[1])
axs.set_ylabel('Count')
axs.tick_params(axis='x', rotation=90) 
plt.tight_layout()
plt.show()
```

```python
data_grouped = df.groupby('States/UTs')['Attempt to commit Murder'].sum().reset_index()

threshold = 0.02  

total = data_grouped['Attempt to commit Murder'].sum()

data_grouped['Percentage'] = data_grouped['Attempt to commit Murder'] / total
data_grouped_sorted = data_grouped.sort_values(by='Attempt to commit Murder', ascending=False)

major_data = data_grouped_sorted[data_grouped_sorted['Percentage'] > threshold]
minor_data = data_grouped_sorted[data_grouped_sorted['Percentage'] <= threshold]

other_data = minor_data[['Attempt to commit Murder']].sum()
other_data = pd.DataFrame({'States/UTs': ['Other'], 'Attempt to commit Murder': other_data})

final_data = pd.concat([major_data, other_data], ignore_index=True)

plt.figure(figsize=(15, 19))
plt.pie(
    final_data['Attempt to commit Murder'], 
    labels=final_data['States/UTs'], 
    autopct='%1.1f%%', 
    colors=plt.cm.Paired(range(len(final_data['States/UTs']))),
    startangle=150,
    wedgeprops={'edgecolor': 'black'}
)

plt.title('Distribution of Attempts to Commit Murder by States/UTs')

plt.show()
```
